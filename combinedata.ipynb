```python

activity=['clap','wave','circle','push']
position = ['i','ii','iii','iv','v','vi','vii','viii','ix']
train_dfs=[]
test_dfs=[]
train_dfs=[]
for gesture in activity:
    for pos in position:
        for k in range(1,11):
          s=f"/content/drive/MyDrive/real data v1/{gesture}/{gesture}_{pos}{k}.csv"
          df = pd.read_csv(s)
          df['label']=gesture
          train_dfs.append(df)

for gesture in activity:
      for k in range(1,6):
          s=f"/content/drive/MyDrive/real data v1/test data/test_{gesture}{k}.csv"
          df = pd.read_csv(s)
          df['label']=gesture
          test_dfs.append(df)

train_df = pd.concat(train_dfs, ignore_index=True)
test_df = pd.concat(test_dfs, ignore_index=True)
```

```python
train_df.shape, test_df.shape
```

```python
train_df.shape[0]+test_df.shape[0]
```

```python
train_label_counts = train_df['label'].value_counts()
test_label_counts = test_df['label'].value_counts()

print(train_label_counts)
print(test_label_counts)
```

```python
# Aligning test_label_counts with train_label_counts
test_label_counts = test_label_counts.reindex(train_label_counts.index).fillna(0)
```

```python
import matplotlib.pyplot as plt

# Define the x-axis locations for the groups
x = range(len(train_label_counts))

# Colors for sophistication
test_data_color = '#c37272'
train_data_color = '#558fd0'

# Bar width
bar_width = 0.4  # Narrower bars

# Plotting test data on the bottom with a sophisticated color
plt.bar(x, test_label_counts, color=test_data_color, width=bar_width, label='Test Data')

# Plotting train data on top of test data with a sophisticated color
plt.bar(x, train_label_counts, bottom=test_label_counts, color=train_data_color, width=bar_width, label='Train Data')

# Adding labels and title
plt.xlabel('Labels')
plt.ylabel('Number of Samples')
plt.title('Raw Training and Testing Data Samples per Label')
plt.xticks(x, train_label_counts.index)  # Set text labels on the x-axis

# Adding a legend
plt.legend(ncol=2, fontsize='small')

# Show the plot
plt.show()
```
